# **Extracción de datos**

Los datos de cada tienda están disponibles en archivos CSV y se cargarán utilizando el código base ya proporcionado. El código extrae información de las cuatro tiendas y organiza los datos en DataFrames utilizando la biblioteca Pandas.

Código base

Detalles importantes:

Código base: Ya está listo para extraer y organizar los datos correctamente, recomendamos no modificarlo.

Los datos se almacenan en DataFrames: cada tienda tiene su propio conjunto de datos (store1, store2, store3 y store4), y debes usarlos para los siguientes pasos del desafío.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
# Para ver la cantidad de filas, columnas, tipos de datos y si hay valores nulos
tienda.info()

# Para ver una descripción estadística básica de las columnas numéricas (promedios, mínimos, máximos)
tienda.describe()
```

# **Ingreso total por cada tienda**

```python
# Sumamos la columna 'Precio' de cada DataFrame
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Imprimimos los resultados formateados para que sean fáciles de leer
print(f"Ingreso Total Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso Total Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso Total Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso Total Tienda 4: ${ingreso_tienda4:,.2f}")
```

# **Ventas por categoría**

```python
# Calculamos la cantidad de ventas por categoría para cada tienda
ventas_cat_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_cat_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_cat_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_cat_tienda4 = tienda4['Categoría del Producto'].value_counts()

# Mostramos los resultados
print("--- Ventas por Categoría: Tienda 1 ---")
print(ventas_cat_tienda1)
print("\n--- Ventas por Categoría: Tienda 2 ---")
print(ventas_cat_tienda2)
print("\n--- Ventas por Categoría: Tienda 3 ---")
print(ventas_cat_tienda3)
print("\n--- Ventas por Categoría: Tienda 4 ---")
print(ventas_cat_tienda4)
```

# **Valoración media por tienda**

```python
# Calculamos la calificación promedio de cada tienda
promedio_tienda1 = tienda['Calificación'].mean()
promedio_tienda2 = tienda2['Calificación'].mean()
promedio_tienda3 = tienda3['Calificación'].mean()
promedio_tienda4 = tienda4['Calificación'].mean()

# Mostramos los resultados redondeados a 2 decimales
print(f"Calificación promedio Tienda 1: {promedio_tienda1:.2f} de 5")
print(f"Calificación promedio Tienda 2: {promedio_tienda2:.2f} de 5")
print(f"Calificación promedio Tienda 3: {promedio_tienda3:.2f} de 5")
print(f"Calificación promedio Tienda 4: {promedio_tienda4:.2f} de 5")
```

# **Productos más vendidos y menos vendidos**

```python
import matplotlib.pyplot as plt

# Agrupamos los DataFrames y sus nombres para recorrerlos fácilmente
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, df in enumerate(tiendas):
    # Contamos las ventas por producto
    ventas_producto = df['Producto'].value_counts()

    # Mostramos los resultados en texto (Top 3 y Bottom 3)
    print(f"--- {nombres[i]} ---")
    print("MÁS vendidos:")
    print(ventas_producto.head(3))
    print("\nMENOS vendidos:")
    print(ventas_producto.tail(3))
    print("-" * 30 + "\n")

    # Creamos la visualización de los 5 productos más vendidos
    plt.figure(figsize=(8, 4))
    ventas_producto.head(5).plot(kind='bar', color='coral')

    plt.title(f'Top 5 Productos Más Vendidos - {nombres[i]}')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad de Ventas')
    plt.xticks(rotation=45, ha='right') # Rota las etiquetas para que se lean bien

    plt.tight_layout() # Ajusta los márgenes
    plt.show() # Muestra el gráfico
```

# **Valor del envío promedio por tienda**

```python
# Calculamos el costo de envío promedio para cada tienda
envio_promedio_tienda1 = tienda['Costo de envío'].mean()
envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

# Imprimimos los resultados con formato de moneda y dos decimales
print(f"Costo de envío promedio Tienda 1: ${envio_promedio_tienda1:,.2f}")
print(f"Costo de envío promedio Tienda 2: ${envio_promedio_tienda2:,.2f}")
print(f"Costo de envío promedio Tienda 3: ${envio_promedio_tienda3:,.2f}")
print(f"Costo de envío promedio Tienda 4: ${envio_promedio_tienda4:,.2f}")
```

# **Generando gráfico**

```python
import matplotlib.pyplot as plt
import pandas as pd

# 0. Preparando los datos para graficar
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos_totales = [tienda['Precio'].sum(), tienda2['Precio'].sum(), tienda3['Precio'].sum(), tienda4['Precio'].sum()]
calificaciones_promedio = [tienda['Calificación'].mean(), tienda2['Calificación'].mean(), tienda3['Calificación'].mean(), tienda4['Calificación'].mean()]

# Unimos todas las tiendas para ver las categorías a nivel global
df_todas = pd.concat([tienda, tienda2, tienda3, tienda4])
categorias_globales = df_todas['Categoría del Producto'].value_counts()

# --- GRÁFICO 1: Gráfico de Barras (Ingresos) ---
plt.figure(figsize=(10, 5))
plt.bar(nombres_tiendas, ingresos_totales, color=['#4C72B0', '#DD8452', '#55A868', '#C44E52'])
plt.title('Ingresos Totales por Tienda', fontsize=14)
plt.ylabel('Ingreso ($)')
plt.show()

# --- GRÁFICO 2: Gráfico Circular (Categorías) ---
plt.figure(figsize=(8, 8))
# Definimos unos colores pastel manualmente para evitar el error
colores_pastel = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0', '#ffb3e6', '#c4e17f', '#76D7C4']

# Reemplazamos cmap por colors=colores_pastel
plt.pie(categorias_globales, labels=categorias_globales.index, autopct='%1.1f%%', colors=colores_pastel, startangle=140)
plt.title('Distribución Global de Ventas por Categoría', fontsize=14)
plt.show()

# --- GRÁFICO 3: Gráfico de Líneas (Calificaciones) ---
plt.figure(figsize=(10, 5))
plt.plot(nombres_tiendas, calificaciones_promedio, marker='o', linestyle='-', color='purple', markersize=10, linewidth=2)
plt.title('Calificación Promedio por Tienda', fontsize=14)
plt.ylabel('Calificación (1 a 5)')
plt.ylim(0, 5.5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# **Informe final**

1. Introducción
El presente informe tiene como objetivo analizar el rendimiento de las cuatro sucursales de la cadena Alura Store para recomendar al Sr. Juan qué tienda debería vender y así poder financiar su nuevo emprendimiento. Para tomar esta decisión, se evaluaron métricas clave: ingresos totales, comportamiento de categorías y productos, satisfacción del cliente (calificaciones) y los costos de envío promedio que asumen los clientes.

2. Desarrollo y Análisis de Datos
A partir de la extracción y visualización de los datos, identificamos los siguientes hallazgos principales:

* Ingresos Totales: Existe una diferencia clara en la rentabilidad. La Tienda 1 es la líder indiscutible, generando la mayor facturación con `$1,150,880,400`. Por el contrario, la Tienda 4 reporta los ingresos más bajos de toda la cadena con `$1,038,375,700`, una diferencia de más de 112 millones respecto a la mejor tienda.

* Categorías y Productos: De manera global, la categoría de Muebles es el motor principal de ventas en todas las tiendas, mientras que los Libros y los Artículos para el hogar son las categorías menos movidas.

* Calificación Promedio `Satisfacción`: La Tienda 3 tiene a los clientes más felices con una puntuación de 4.05 / 5.0. Curiosamente, la Tienda 1 (la que más vende) tiene la peor calificación promedio con 3.98 / 5.0, lo que indica que venden mucho, pero tienen áreas de oportunidad en su servicio.

* Costo de Envío Promedio: Sabiendo que este costo lo asume el cliente, notamos que la Tienda 1 tiene los envíos más caros en promedio `($26,018)`, lo cual podría explicar su baja calificación. La Tienda 4, a pesar de tener las ventas más bajas, ofrece los envíos más económicos `($23,459)`.

3. Conclusión y Recomendación Final
Basado estrictamente en la eficiencia financiera y el rendimiento comercial, la recomendación para el Sr. Juan es vender la Tienda 4.

Justificación:

Aunque la Tienda 4 tiene el costo de envío más amigable para el usuario y mantiene una calificación aceptable (4.00), es la tienda menos rentable de toda la cadena, posicionándose en el último lugar de ingresos. Mantener una sucursal con el volumen de ventas más bajo limita el flujo de caja del negocio global.

Vender la Tienda 4 le permitirá al Sr. Juan obtener el capital necesario para su nuevo emprendimiento sin sacrificar a su "gallina de los huevos de oro" (Tienda 1), ni a sus tiendas con mejor retención y satisfacción de clientes (Tiendas 2 y 3). Se sugiere, además, que el Sr. Juan utilice parte de su tiempo en el nuevo emprendimiento para auditar la logística de la Tienda 1 y bajar esos altos costos de envío que están afectando su calificación.
